```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv"
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()

# Group by 'drive-wheels' and 'body-style', and count occurrences
df_group_count = df.groupby(['drive-wheels', 'body-style']).size().reset_index(name='counts')
print(df_group_count)

# Pivot table to organize the counts
grouped_pivot_counts = df_group_count.pivot(index='drive-wheels', columns='body-style', values='counts')
grouped_pivot_counts = grouped_pivot_counts.fillna(0)
print(grouped_pivot_counts)

# Visualize the counts with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(grouped_pivot_counts, annot=True, fmt="g", cmap='RdBu', linewidths=.5)
plt.title('Heatmap of Drive-Wheels vs. Body-Style Counts')
plt.show()

# Boxplot for body-style and price
plt.figure(figsize=(10, 8))
sns.boxplot(x="body-style", y="price", data=df)
plt.title('Boxplot of Body-Style vs. Price')
plt.show()

# Boxplot for drive-wheels and price
plt.figure(figsize=(10, 8))
sns.boxplot(x="drive-wheels", y="price", data=df)
plt.title('Boxplot of Drive-Wheels vs. Price')
plt.show()
```
